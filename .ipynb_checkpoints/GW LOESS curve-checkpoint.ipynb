```python
###Script found on github - SOURCE!

def lowess(x, y, f=2. / 3., iter=3):
    """lowess(x, y, f=2./3., iter=3) -> yest
    Lowess smoother: Robust locally weighted regression.
    The lowess function fits a nonparametric regression curve to a scatterplot.
    The arrays x and y contain an equal number of elements; each pair
    (x[i], y[i]) defines a data point in the scatterplot. The function returns
    the estimated (smooth) values of y.
    The smoothing span is given by f. A larger value for f will result in a
    smoother curve. The number of robustifying iterations is given by iter. The
    function will run faster with a smaller number of iterations.
    """
    
    x, y = zip(*sorted(zip(x, y)))
    x=np.array(x)
    y=np.array(y)
    
    n = len(x)
    r = int(ceil(f * n))
    h = [np.sort(np.abs(x - x[i]))[r] for i in range(n)]
    w = np.clip(np.abs((x[:, None] - x[None, :]) / h), 0.0, 1.0)
    w = (1 - w ** 3) ** 3
    yest = np.zeros(n)
    delta = np.ones(n)
    for iteration in range(iter):
        for i in range(n):
            weights = delta * w[:, i]
            b = np.array([np.sum(weights * y), np.sum(weights * y * x)])
            A = np.array([[np.sum(weights), np.sum(weights * x)],
                          [np.sum(weights * x), np.sum(weights * x * x)]])
            beta = linalg.solve(A, b)
            yest[i] = beta[0] + beta[1] * x[i]

        residuals = y - yest
        s = np.median(np.abs(residuals))
        delta = np.clip(residuals / (6.0 * s), -1, 1)
        delta = (1 - delta ** 2) ** 2

    return yest
```

```python
def lowessGW(data, xv, yv, east, north, xcoord, ycoord, d, k = 'gaussian', f=0.66, iter=3):
    """lowess(x, y, f=2./3., iter=3) -> yest
    Lowess smoother: Robust locally weighted regression.
    The lowess function fits a nonparametric regression curve to a scatterplot.
    The arrays xv and yv contain an equal number of elements; each pair
    (xv[i], yv[i]) defines a data point in the scatterplot. The function returns
    the estimated (smooth) values of y.
    East, North, dist represent the origin of the location to weight the GWloess cruve to. 
    d represents the bandwidth to use in the distance weighting function.
    k represents the required kernal. Choices are currently limited to guassian, bi-squared and discrete.
    data is dataframe so gw weighting can be calculated
    The smoothing span is given by f. A larger value for f will result in a
    smoother curve. The number of robustifying iterations is given by iter. The
    function will run faster with a smaller number of iterations.
    """
 

    data['dist'] = ((xcoord - data[east])**2 + (ycoord - data[north])**2) ** 0.5
    
    if k == "gaussian":
        gw_max = np.sqrt(np.log(0.0000001)/-0.5)*d
        data_s = data[data['dist']<=gw_max]
        data_s['gw'] = np.exp(-0.5 * (data_s.dist/d)**2)
    elif k == "bi-square":
        data_s = data[data['dist']<=d]
        data_s['gw'] = (1-((data_s.dist/d)**2))**2
    else:
        data_s = data[data['dist']<=d]
        data_s['gw'] = np.where(data_s['dist'] <= d, 1, 0)    
    #print(data.head())
    
    
    x, y , g = zip(*sorted(zip(data_s[xv], data_s[yv], data_s.gw)))
    x=np.array(x)
    y=np.array(y)
    g=np.array(g)
    
    n = len(x)
    r = int(ceil(f * n))
    h = [np.sort(np.abs(x - x[i]))[r] for i in range(n)]
    t = np.clip(np.abs((x[:, None] - x[None, :]) / h), 0.0, 1.0)
    t = (1 - t ** 3) ** 3

    w = t * g[:, None]

    
    
    yest = np.zeros(n)
    delta = np.ones(n)
    for iteration in range(iter):
        for i in range(n):
            weights = delta * w[:, i]
            b = np.array([np.sum(weights * y), np.sum(weights * y * x)])
            A = np.array([[np.sum(weights), np.sum(weights * x)],
                          [np.sum(weights * x), np.sum(weights * x * x)]])
            beta = linalg.solve(A, b)
            yest[i] = beta[0] + beta[1] * x[i]

        residuals = y - yest
        s = np.median(np.abs(residuals))
        delta = np.clip(residuals / (6.0 * s), -1, 1)
        delta = (1 - delta ** 2) ** 2

    frame = {'Date':data_s.first_listing_Date , 'xx': data_s.xx,  'LOESS': yest }
    results = pd.DataFrame(frame)
    
    return results
```

```python
#set up SQL connection
conn = pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};' 
                      'Server=NE-SQL-CL1-03.kf-ext.com;' 
                      'Trusted_Connection=yes;')
cursor = conn.cursor()

listSQl = """ Select [DayOfYearID] + ((year([first_listing_Date])-2015)*365) as xx
	  ,street
      ,[postcode]
      ,[beds]
	  ,bb.branch_name
      ,[first_listing_price] as price
      ,[first_listing_Date]
	  ,cc.oseast1m * 1e0 as Easting
	  ,cc.osnrth1m * 1e0 as Northing
      ,company_name
  FROM [Rightmove].[dbo].[PropertyRent]aa
  inner join
  [Rightmove].[dbo].[DimAgent]bb
  on aa.[agent_id] = bb.[agent_id]
  inner join
  postcode.dbo.onspostcodedirectory cc
  on aa.postcode = cc.pcds
  inner join
  DWStaging.cdm.Dim_Date dd
  on aa.first_listing_date = dd.DateValue
  where first_listing_date >= '01 jan 2018'
  and flag_delete = 0
  and isnull([short_let_flag],0) <> 1
  and isnull([price_per_person_flag],0) <> 1
  and new_build = 0
  and oslaua = 'E09000024'
  and property_type_id = 2
  and beds > 0 and beds < 3
  AND oseast1m > 0
  and first_listing_price < 5000
  and first_listing_price > 500
  order by [DateID]
"""


listing = pd.read_sql(listSQl,conn)
conn.close()
```

```python
#### Test Merton LOESS against two different locations within the borough - Wimbledon Village and Morden
%%time 
RentSW194 = lowessGW(data = listing, xv = 'xx', yv = 'price', east = 'Easting', north = 'Northing', xcoord = 523924, ycoord = 170729, d = 250, k = 'gaussian', f=0.2, iter=1)
RentSM45 = lowessGW(data = listing, xv = 'xx', yv = 'price', east = 'Easting', north = 'Northing', xcoord = 525662, ycoord = 168059, d = 250, k = 'gaussian', f=0.2, iter=1)
Rentest = lowess(listing.xx, listing.price, f=0.2, iter=2)
```

```python

plt.scatter(listing.xx,listing.price,2,'lightgrey')
plt.plot(listing.xx,Rentest,color='tomato')
plt.plot(RentSW194.xx,RentSW194.LOESS,color='green')
plt.plot(RentSM45.xx,RentSM45.LOESS,color='blue')
```
